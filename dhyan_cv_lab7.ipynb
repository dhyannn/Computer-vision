```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
from google.colab.patches import cv2_imshow

# Read the original image
original_image = cv2.imread('/content/park.jpg', cv2.IMREAD_GRAYSCALE)

# Display the original image
plt.subplot(1, 2, 1)
plt.imshow(original_image, cmap='gray')
plt.title('Original Image')
plt.xticks([]), plt.yticks([])

# Add Gaussian noise to the original image
mean = 0
variance = 100
sigma = variance ** 0.5
gaussian_noise = np.random.normal(mean, sigma, original_image.shape)
noisy_image = original_image + gaussian_noise

# Display the noisy image
plt.subplot(1, 2, 2)
plt.imshow(noisy_image, cmap='gray')
plt.title('Noisy Image')
plt.xticks([]), plt.yticks([])

plt.show()

# Denoising the image using Gaussian blur
denoised_image = cv2.GaussianBlur(noisy_image, (5, 5), 0)

# Display the denoised image
plt.imshow(denoised_image, cmap='gray')
plt.title('Denoised Image')
plt.xticks([]), plt.yticks([])
plt.show()
```
