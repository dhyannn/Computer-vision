```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
from google.colab.patches import cv2_imshow
```

```python
image = cv2.imread('/content/cat.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
# Display the original image
plt.subplot(2, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.xticks([]), plt.yticks([])
```

```python
# First Order Derivative Methods (Sobel)
sobel_x = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=3)
sobel_y = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=3)
sobel_combined = cv2.magnitude(sobel_x, sobel_y)

plt.subplot(2, 2, 2)
plt.imshow(sobel_combined, cmap='gray')
plt.title('First Order Derivative (Sobel)')
plt.xticks([]), plt.yticks([])
```

```python
# Second Order Derivative Methods (Laplacian)
laplacian = cv2.Laplacian(image, cv2.CV_64F)

plt.subplot(2, 2, 3)
plt.imshow(laplacian, cmap='gray')
plt.title('Second Order Derivative (Laplacian)')
plt.xticks([]), plt.yticks([])
```

```python
# Optimum Edge Detection Method (Canny)
canny = cv2.Canny(image, 100, 200)

plt.subplot(2, 2, 4)
plt.imshow(canny, cmap='gray')
plt.title('Optimum Edge Detection (Canny)')
plt.xticks([]), plt.yticks([])

plt.tight_layout()
plt.show()
```
